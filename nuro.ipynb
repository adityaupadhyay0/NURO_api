### 📊 The 'Big Three' Neuromarketing Metrics for A/B Testing

When testing two versions of an ad, these are the three neural signatures that predict real-world success:

#### 1. Visual Attention (The 'Stop the Scroll' Metric)
*   **Brain Region:** V1 (Primary Visual Cortex) and Parietal Lobe.
*   **What it means:** Does the ad physically capture the viewer's eyes? High activity here means your visual creative is 'stopping the scroll.' If this is low, your audience might be looking away or ignoring the ad entirely.

#### 2. Emotional Engagement / Liking (The 'Brand Affinity' Metric)
*   **Brain Region:** Insula and Ventral Striatum (Reward Centers).
*   **What it means:** Is the viewer having a positive emotional reaction? This is the best predictor of whether a customer will feel good about your brand after the ad ends. A positive 'Liking' score means the creative is building 'Brand Love.'

#### 3. Memory Encoding (The 'Purchase Intent' Metric)
*   **Brain Region:** Parahippocampal Gyrus / Hippocampus.
*   **What it means:** Is the brain actually 'saving' this information? This is the most important metric for long-term ROI. If they don't encode the memory, they won't remember your brand name when they are standing in the store aisle later.

**PRO TIP:** The best ads usually have a **'Peak-End'** pattern: a strong 'Peak' of Attention in the middle to keep them interested, and a high 'Memory' score at the very end when the logo appears.

### 🧠 Neuroscience Lesson: How the Brain Processes Multimodal Input

TRIBE v2 is a **Foundation Model for In-Silico Neuroscience**, meaning it is designed to act like a digital twin of the human cortex. Here are three key concepts it simulates:

#### 1. Functional Specialization (The Map)
Just like a city has different districts, your brain has specific regions for specific tasks.
* **Occipital Lobe (Back of the brain):** Primarily handles vision. The model uses a 'Video Encoder' (V-JEPA) to mimic how your **V1 (Primary Visual Cortex)** detects edges and motion.
* **Temporal Lobe (Sides of the brain):** Handles hearing. The model uses an 'Audio Encoder' (Wav2Vec) to mimic your **A1 (Primary Auditory Cortex)**.

#### 2. Hierarchical Processing
Information doesn't just 'arrive' in the brain fully formed. It travels up a ladder:
1. **Primary Areas (V1/A1):** Detect raw features like brightness or pitch.
2. **Association Areas:** Combine these features into objects (a face) or sounds (a word).
3. **Frontal/Parietal Areas:** Interpret the *meaning* of what you see and hear.

#### 3. Multimodal Integration
This is the most advanced part! Your brain has 'convergence zones' where sight and sound meet. For example, if you see someone's lips move, your visual cortex sends signals to your auditory cortex to help you 'hear' the words better (the McGurk Effect).

**TRIBE v2 uses a 'Transformer' architecture to simulate this integration, allowing 'visual' neurons to talk to 'auditory' neurons to predict a unified brain response.**

```python
import numpy as np
import pandas as pd
from tribev2 import TribeModel
from nilearn import datasets
from huggingface_hub import login
import json

class NeuroPersuasionAPI:
    def __init__(self, hf_token, cache_dir="./cache"):
        """Initialize the SaaS engine with HF authentication."""
        login(token=hf_token.strip())
        self.cache_dir = cache_dir
        self.model = TribeModel.from_pretrained("facebook/tribev2", cache_folder=cache_dir)
        self.atlas = datasets.fetch_atlas_surf_destrieux()
        self.left_map = self.atlas['map_left']

        # Standardized ROI Mappings for SaaS Metrics
        self.metrics_config = {
            'attention': 52,    # Superior Parietal
            'engagement': 18,   # Insula
            'memory': 34,       # Parahippocampal
            'social': 65        # Superior Temporal
        }

    def _get_roi_signal(self, preds, region_idx):
        half = preds.shape[1] // 2
        mask = (self.left_map == region_idx)
        return preds[:, :half][:, mask].mean(axis=1)

    def analyze_video(self, video_path):
        """Primary API Endpoint for video processing."""
        # 1. Inference
        df_events = self.model.get_events_dataframe(video_path=video_path)
        preds, _ = self.model.predict(events=df_events)

        # 2. Extract Temporal Signals
        signals = {name: self._get_roi_signal(preds, idx) for name, idx in self.metrics_config.items()}

        # 3. Calculate NEI (Neural Engagement Index)
        sig_array = np.array(list(signals.values()))
        corr_matrix = np.corrcoef(sig_array)
        nei = corr_matrix[np.triu_indices(len(signals), 1)].mean()

        # 4. Construct JSON Response
        response = {
            "status": "success",
            "metadata": {"frames": len(preds), "vertices": preds.shape[1]},
            "aggregate_scores": {
                name: float(sig.mean()) for name, sig in signals.items()
            },
            "neural_engagement_index": float(nei),
            "verdict": "High Impact" if nei > 0.05 else "Low Impact"
        }
        return response

print("NeuroPersuasionAPI Class defined and ready for integration.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tribev2 import TribeModel
from nilearn import datasets
from huggingface_hub import login
import json
from datetime import datetime

class NeuroEngineV3:
    """The Elite SaaS Engine for Neuromarketing Intelligence."""
    def __init__(self, hf_token):
        login(token=hf_token.strip())
        self.model = TribeModel.from_pretrained("facebook/tribev2", cache_folder="./cache")
        self.destrieux = datasets.fetch_atlas_surf_destrieux()
        self.left_atlas = self.destrieux['map_left']

        self.kpi_map = {
            'Visual Saliency (Attention)': 52,
            'Emotional Resonance (Liking)': 18,
            'Memory Encoding (Retention)': 34,
            'Story Arc (Narrative)': 65,
            'Cognitive Efficiency': 24
        }

    def _extract_high_res_signals(self, preds):
        half = preds.shape[1] // 2
        return {name: preds[:, :half][:, (self.left_atlas == idx)].mean(axis=1)
                for name, idx in self.kpi_map.items()}

    def analyze_asset(self, video_path, asset_name="Asset_01"):
        df_events = self.model.get_events_dataframe(video_path=video_path)
        preds, _ = self.model.predict(events=df_events)
        signals = self._extract_high_res_signals(preds)

        # Advanced Metrics
        sig_matrix = np.array(list(signals.values()))
        nei = np.corrcoef(sig_matrix)[np.triu_indices(len(signals), 1)].mean()

        return {
            "id": f"NE-{datetime.now().strftime('%Y%m%d%H%M')}",
            "name": asset_name,
            "scores": {k: float(v.mean()) for k, v in signals.items()},
            "nei": float(nei),
            "raw_curves": signals
        }

    def run_competitive_audit(self, path_a, path_b):
        """Runs a side-by-side A/B test with deep neurological benchmarking."""
        print("🚀 Initializing Neural Audit: Comparison Mode...")
        report_a = self.analyze_asset(path_a, "Version A")
        report_b = self.analyze_asset(path_b, "Version B")

        # Determine Winner
        wins_a = sum(1 for k in self.kpi_map if report_a['scores'][k] > report_b['scores'][k])
        winner = "Version A" if wins_a > 2 else "Version B"

        return {"A": report_a, "B": report_b, "winner": winner, "confidence": f"{(wins_a/5)*100}%"}

print("NeuroEngine V3 Elite: Compiled and Secure.")
```

```python
def render_saas_dashboard(audit_results):
    """Generates a client-ready UI Report."""
    plt.style.use('dark_background')
    fig, axes = plt.subplots(3, 1, figsize=(16, 18))
    colors = ['#00d1ff', '#ff007a', '#00ff88', '#f39c12', '#9b59b6']

    # 1. Comparison Bar Chart
    df_plot = pd.DataFrame({
        "Metric": list(audit_results['A']['scores'].keys()),
        "Version A": list(audit_results['A']['scores'].values()),
        "Version B": list(audit_results['B']['scores'].values())
    }).set_index("Metric")
    df_plot.plot(kind='barh', ax=axes[0], color=['#444444', '#00d1ff'])
    axes[0].set_title("NEURO-METRIC BENCHMARK", fontsize=20, pad=20)

    # 2. Neural Pulse (Attention Over Time)
    axes[1].plot(audit_results['A']['raw_curves']['Visual Saliency (Attention)'], label='A', color='gray', alpha=0.5)
    axes[1].plot(audit_results['B']['raw_curves']['Visual Saliency (Attention)'], label='B', color='#00d1ff', lw=3)
    axes[1].set_title("REAL-TIME ATTENTION SPECTRUM", fontsize=16)
    axes[1].legend()

    # 3. Synchrony Heatmap (Version B - Winner)
    corr = np.corrcoef(np.array(list(audit_results['B']['raw_curves'].values())))
    sns.heatmap(corr, annot=True, xticklabels=list(audit_results['B']['scores'].keys()),
                yticklabels=list(audit_results['B']['scores'].keys()), ax=axes[2], cmap='mako')
    axes[2].set_title("WINNING VERSION: NETWORK COHERENCE (NEI)", fontsize=16)

    plt.tight_layout()
    plt.show()

    print(f"\n{'='*60}\n FINAL SCIENTIFIC VERDICT: {audit_results['winner']} IS SUPERIOR\n{'='*60}")
```
